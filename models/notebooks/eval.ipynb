```python
from sklearn.metrics.cluster import adjusted_rand_score
import pandas as pd
```

```python
df = pd.read_csv('../data/cluster_label_for_dataset.csv')
df.sample(3)
```

```python
res = [
    ['kmeans', 'kmeans', 'birch'],
    ['birch', 'gaussian', 'gaussian'],
    [adjusted_rand_score(df['kmeans_label'].values, df['birch_label'].values),
    adjusted_rand_score(df['kmeans_label'].values, df['gaussian_label'].values),
    adjusted_rand_score(df['birch_label'].values, df['gaussian_label'].values)]
]
pd.DataFrame(res).T
```
