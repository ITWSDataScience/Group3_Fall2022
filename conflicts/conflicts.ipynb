```python
import pandas as pd

df = pd.read_csv('icews_conflict_data_2019_2023.csv')
```

```python
countries = None
with open('countries.txt', 'r') as f:
    countries = f.readlines()
countries = [x.strip() for x in countries]
countries[:5]
```

```python
africa_only = df[df['Target Country'].isin(countries)]
africa_only
```

```python
africa_only['Target Country'].value_counts()
```

```python
africa_only.to_csv('african_conflicts.csv', index=False)
```
