```python
import csv
import json
```

```python
eqreader = None
disasters = []

with open('./earthquakes/earthquakes.tsv', 'r') as f:
    eqreader = csv.reader(f, delimiter='\t', quotechar='"')

    for row in eqreader:
        if row[0] != 'Search Parameters': 
            if int(row[1]) >= 2021:
                info = f'{row[1]}-{row[2]}-{row[3]} {row[4]}:{row[5]}:{row[6]} Location Name: {row[9]}'
                lat = float(row[10])
                lon = float(row[11])
            
                disasters.append({
                    'info': info,
                    'lat': lat,
                    'lon': lon,
                    'dtype': 'earthquake'
                })
```

```python
with open('./tsunamis/runups.tsv', 'r') as f:
    eqreader = csv.reader(f, delimiter='\t', quotechar='"')

    i = 0
    for row in eqreader:
        if i < 2:
            i += 1
            continue # skip predata
        
        if len(row[17]) == 0 or len(row[18]) == 0:
            continue
        if int(row[1]) >= 2021:
            info = f'{row[1]}-{row[2]}-{row[3]} {row[4]}:{row[5]}:{row[6]} Location Name: {row[11]}'
            lat = float(row[17])
            lon = float(row[18])
            
            disasters.append({
                    'info': info,
                    'lat': lat,
                    'lon': lon,
                    'dtype': 'tsunami'
            })
```

```python
with open('disasters.json', 'w', encoding='utf-8') as f:
    json.dump(disasters, f, ensure_ascii=False, indent=4)
```

```python
import pandas as pd
```

```python
df = pd.read_csv('./hurricanes/hurricanes.csv')
```

```python
df.drop(index=df.index[0], axis=0, inplace=True)
```

```python
df = df.astype({'SEASON': 'int'})
```

```python
df.head()
```

```python
df = df[df['SEASON'] >= 2021]
```

```python
hurricanes = []
sids = df['SID'].unique()
g = df.groupby(df['SID'])

for sid in sids:
    if sid != ' ':
        gdf = g.get_group(sid)

        path = []
        for i, row in gdf.iterrows():
            path.append([float(row['LAT']), float(row['LON'])])
        
        hurricanes.append({
            'sid': sid,
            'path': path
        })
```

```python
with open('hurricanes.json', 'w', encoding='utf-8') as f:
    json.dump(hurricanes, f, ensure_ascii=False, indent=4)
```
